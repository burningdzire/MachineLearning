```python
#Plotting Univariate Plots
#1. Histogram
sns.distplot(df['Age'], hist = True, kde = True, rug = False)
```

```python
plt.show()
```

```python
sns.distplot(df['Salary'], kde = True, rug = False)
```

```python
plt.show()
```

```python
plt.close()
```

```python
#Kernel Density Estimation(kde)
sns.kdeplot(df['Age'], shade = True)
plt.show()
```

```python
plt.close()
```

```python
plt.subplot(211)
sns.kdeplot(df['Salary'], bw = 500, label = 'bin size = 500')
plt.subplot(212)
sns.kdeplot(df['Salary'], bw = 100, label = 'bin size = 100')
plt.legend();
plt.show()
```

```python
#Plotting Bivariate Plots
# Scatterplot
sns.jointplot(x = 'ID', y = 'Salary', data = df)
plt.show()
```

```python
plt.close()
```

```python
#Kernel Density Estimation for Bivariate Data
sns.jointplot(x = df['Age'], y = df['Salary'], kind = 'hex', color = 'blue')
plt.show()
```

```python
plt.close()
```
