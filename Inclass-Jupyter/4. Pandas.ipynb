```python
#Univariate data plotting
#histogram
sns.distplot(iris['sepal_length'])
plt.show()
```

```python
plt.close()
```

```python
plt.subplot(221)
sns.distplot(iris['sepal_length'])
plt.subplot(222)
sns.distplot(iris['sepal_width'])
plt.subplot(223)
sns.distplot(iris['petal_length'])
plt.subplot(224)
sns.distplot(iris['petal_width'])
plt.show()
```

```python
#Scatter plot of iris dataset
sns.jointplot(x = 'sepal_length', y = 'petal_length', data = iris)
plt.show()
```

```python
#KDE curve
sns.jointplot(x = iris['petal_length'], y = iris['petal_width'], kind = 'kde', color = 'blue')
plt.show()
```

```python
#Stripplot
sns.stripplot(x = 'species', y = 'sepal_length', data = iris)
plt.show()
```
