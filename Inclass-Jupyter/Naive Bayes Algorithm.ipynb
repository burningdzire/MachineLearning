```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df['sex'] = np.where(df['sex'] == 'male', 0, 1)
df.head()
```

```python
# from sklearn.preprocessing import LabelEncoder
# le = LabelEncoder()
# le.fit_transform(df['embarked'])

df['embarked'] = np.where(df['embarked'] == 'S', 0, np.where(df['embarked'] == 'C', 1, np.where(df['embarked'] == 'R', 2, 3)))
df.head()
```

```python
df = df[['survived', 'pclass', 'sex', 'age', 'parch', 'embarked']].dropna(axis = 0, how = 'any')
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split as tts
train, test = tts(df, test_size = 0.3)
```

```python
xtrain = train.loc[ : , train.columns != 'survived']
xtest = test.loc[ : , test.columns != 'survived']

ytrain = train['survived']
ytest = test['survived']
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(xtrain, ytrain)
```

```python
ypred = gnb.predict(xtest)
```

```python
result = pd.DataFrame({'actual': ytest, 'predicted': ypred})
```
