```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Plotting a simple line
x = [1, 2, 3, 4]
y = [4, 7, 10, 13]
plt.plot(x, y)
```

```python
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('This is a Simple Line')
```

```python
plt.show()
```

```python
plt.close()
```
